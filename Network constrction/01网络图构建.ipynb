###包含SFeature_1, SFeature_2, SFeature_3

```python

from itertools import combinations
import networkx as nx
import pandas as pd


def process_episode(episode_number):
    # Construct the file path for the episode
    file_path = f'D:/研究生论文/小论文2/代码/03网络图/00关联分析_强关联规则/输出结果/03共线关系预备数据/桃花马上请长缨RAG/第{episode_number}集_拆分.xlsx'
    df = pd.read_excel(file_path)

    # 提取质量特征列
    features_list = df[['SFeature_1', 'SFeature_2', 'SFeature_3']].apply(lambda row: row.dropna().tolist(), axis=1)

    # 将提取的质量特征列存储为新列
    df['Merged_Features'] = features_list

    # 确保'Merged_Features'列中的字符串被正确解析为列表
    df['Merged_Features'] = df['Merged_Features'].apply(lambda x: eval(x) if isinstance(x, str) else x)

    # 构建共现网络
    G = nx.Graph()

    # 使用边的权重来记录共现次数
    if len(df['Merged_Features']) >= 2:
        for topics in df['Merged_Features']:
            for pair in combinations(topics, 2):
                if G.has_edge(*pair):
                    G.edges[pair]['weight'] += 1
                else:
                    G.add_edge(*pair, weight=1)
    

    # Export the edges to a CSV file
    edges_df = nx.to_pandas_edgelist(G)
    edges_excel_path = f'D:/研究生论文/小论文2/代码/03网络图/01网络图构建/边数据/桃花马上请长缨RAG/network_edges_episode_{episode_number}.xlsx'
    edges_df.to_excel(edges_excel_path, index=False)
    return edges_excel_path

# Process each episode and save the edges
edges_paths = []
for episode in range(1, 92):  # Assuming there are 10 episodes
    edges_path = process_episode(episode)
    edges_paths.append(edges_path)

edges_paths
```
