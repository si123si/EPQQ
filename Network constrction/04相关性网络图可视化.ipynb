```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# 存储每一集的网络图和节点信息
networks = []
all_nodes = set()

# 循环1到15
for episode in range(1, 3):
    # 读取数据
    df = pd.read_excel(f"E:/桌面/代码/03网络图/02节点之间关系/三体4.8/correlation_results_episode_{episode}_updated.xlsx")
    pagerank_df = pd.read_excel(f"E:/桌面/代码/03网络图/03节点权重/三体4.8/pagerank_results_{episode}_updated.xlsx")

    # 创建无向图
    G = nx.Graph()

    # 添加边和权重
    for _, row in df.iterrows():
        source = row['source']
        target = row['target']
        correlation = row['correlation']
        # 根据correlation的正负确定边的标签
        edge_label = '+' if correlation >= 0 else '-'
        # 添加边
        G.add_edge(source, target, weight=abs(correlation), label=edge_label)

    # 获取节点的PageRank值
    pagerank_dict = dict(zip(pagerank_df['Node Name'], pagerank_df['PageRank']))
    G.graph['pagerank_dict'] = pagerank_dict

    # 添加到网络列表中
    networks.append((G, f'Episode {episode}'))

    # 更新节点集合
    all_nodes.update(G.nodes())

# 创建绘图对象
plt.figure(figsize=(20, 6))

# 设置节点布局
pos = nx.spring_layout(networks[0][0], seed=42)

# 绘制每一集的网络
for i, (G, label) in enumerate(networks):
    # 设置当前子图位置
    plt.subplot(1, len(networks), i + 1)
    
    # 绘制节点，相同节点为相同颜色
    node_colors = [G.graph['pagerank_dict'].get(node, 0) for node in G.nodes()]  # 使用0填充缺失的节点
    nx.draw_networkx_nodes(G, pos, node_size=4000, node_color=node_colors, cmap=plt.cm.Set3, label=label)

    # 绘制边
    for u, v, d in G.edges(data=True):
        nx.draw_networkx_edges(G, pos, edgelist=[(u, v)], width=d['weight']*8, edge_color='black')

# 绘制相同节点之间的弯曲虚线连接
for node in all_nodes:
    # 找出相同节点在不同集中的连接
    connections = [(u, v) for u, v in G.edges() if node in (u, v)]
    for u, v in connections:
        for i, (G, _) in enumerate(networks[1:], start=1):
            if G.has_edge(u, v):
                # 计算节点之间的弯曲虚线路径
                path = nx.shortest_path(G, u, v)
                if len(path) > 2:
                    edge_points = np.array([pos[node] for node in path])
                    nx.draw_networkx_edges(G, pos, edgelist=[(u, v)], style='dashed', alpha=0.5, width=3,
                                           connectionstyle=f'arc3,rad={0.1*i}', edge_color='black')

# 添加节点标签，设置字体为新罗马10号
for G, _ in networks:
    nx.draw_networkx_labels(G, pos, font_family='Times New Roman', font_size=10)

# 显示图像
plt.axis('off')
plt.tight_layout()
plt.show()
```
