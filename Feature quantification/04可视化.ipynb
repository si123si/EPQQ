```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 读取数据
data = pd.read_excel("D:/研究生论文/小论文2/代码/04质量评估_Ising模型/01质量评估/01节点质量/无限世界RAG/all_node_energy_english.xlsx")

# 创建图形
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# 提取数据
node_names = data['Node Name']
episodes = data['Episode']
node_energies = abs(data['Node Energy'])

# 获取唯一的节点名称和集数
unique_node_names = node_names.unique()
unique_episodes = episodes.unique()

# 为每个节点名称和集数分配一个位置
y_pos = [unique_node_names.tolist().index(node) for node in node_names]
x_pos = [unique_episodes.tolist().index(episode) for episode in episodes]

# 设置柱子的宽度和深度
x_width = 0.4
y_width = 0.4

# 绘制柱形图
ax.bar3d(x_pos, y_pos, [0]*len(node_energies), x_width, y_width, node_energies, shade=True)

# 设置坐标轴标签
ax.set_xlabel('Episode', fontname='Times New Roman', fontsize=12)
ax.set_ylabel('Node Name', fontname='Times New Roman', fontsize=12)
ax.set_zlabel('Absolute Node Energy', fontname='Times New Roman', fontsize=12)

# 设置坐标轴刻度标签
ax.set_xticks(range(len(unique_episodes)))
ax.set_xticklabels(unique_episodes, fontname='Times New Roman', fontsize=10)
ax.set_yticks(range(len(unique_node_names)))
ax.set_yticklabels(unique_node_names, rotation=0, fontname='Times New Roman', fontsize=10)

# 设置坐标轴范围
ax.set_xlim(-0.5, len(unique_episodes) - 0.5)
ax.set_ylim(-0.5, len(unique_node_names) - 0.5)
ax.set_zlim(0, max(node_energies) * 1.1)

# 显示图形
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# 读取数据
data = pd.read_excel("D:/研究生论文/小论文2/代码/04质量评估_Ising模型/01质量评估/01节点质量/无限世界RAG/all_node_energy_english.xlsx")

# 计算每个质量特征的绝对值之和
feature_sums = data.groupby('Node Name')['Node Energy'].apply(lambda x: x.abs().sum())

# 获取绝对值和最大的 3 个特征和最小的 3 个特征
top_3_features = feature_sums.nlargest(3).index
bottom_3_features = feature_sums.nsmallest(3).index
selected_features = top_3_features.union(bottom_3_features)

# 过滤数据，只保留选定的特征
filtered_data = data[data['Node Name'].isin(selected_features)]

# 创建图形
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# 提取数据
node_names = filtered_data['Node Name']
episodes = filtered_data['Episode']
node_energies = abs(filtered_data['Node Energy'])

# 获取唯一的节点名称和集数
unique_node_names = node_names.unique()
unique_episodes = sorted(episodes.unique())

# 为每个节点名称和集数分配一个位置
y_pos = [unique_node_names.tolist().index(node) for node in node_names]
x_pos = [unique_episodes.index(episode) for episode in episodes]

# 设置柱子的宽度和深度
x_width = 0.4
y_width = 0.4

# 设置渐变颜色
norm = plt.Normalize(node_energies.min(), node_energies.max())
colors = plt.cm.viridis(norm(node_energies))

# 绘制柱形图
for i in range(len(node_energies)):
    ax.bar3d(x_pos[i], y_pos[i], 0, x_width, y_width, node_energies.iloc[i], color=colors[i])

# 设置坐标轴标签
ax.set_xlabel('Episode', fontname='Times New Roman', fontsize=12)
ax.set_ylabel('Node Name', fontname='Times New Roman', fontsize=12)
ax.set_zlabel('Absolute Node Energy', fontname='Times New Roman', fontsize=12)

# 设置坐标轴刻度标签
ax.set_xticks(range(len(unique_episodes)))
ax.set_xticklabels(unique_episodes, fontname='Times New Roman', fontsize=10)
ax.set_yticks(range(len(unique_node_names)))
ax.set_yticklabels(unique_node_names, rotation=0, fontname='Times New Roman', fontsize=10)

# 设置坐标轴范围
ax.set_xlim(-0.5, len(unique_episodes) - 0.5)
ax.set_ylim(-0.5, len(unique_node_names) - 0.5)
ax.set_zlim(0, max(node_energies) * 1.1)

# 添加颜色条
sm = plt.cm.ScalarMappable(cmap='viridis', norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, shrink=0.5, aspect=5)
cbar.set_label('Absolute Node Energy', fontname='Times New Roman', fontsize=12)

# 设置x轴刻度以2集为间隔
ax.set_xticks(range(0, len(unique_episodes), 2))
ax.set_xticklabels(unique_episodes[::2], fontname='Times New Roman', fontsize=10)

# 显示图形
plt.show()
```
