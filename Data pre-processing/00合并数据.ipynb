###合并数据

##b站动漫数据

```python
import pandas as pd
import os

# 设置包含所有Excel文件的目录路径
file_directory = 'D:/研究生论文/小论文2/代码/01数据预处理/00data-初始-未合并集数/18择天记第二季'   

# 列出所有Excel文件
excel_files = [file for file in os.listdir(file_directory) if file.endswith('.xlsx') or file.endswith('.xls')]

# 初始化一个空的DataFrame来存储所有数据
all_data = pd.DataFrame()

for filename in excel_files:
    #if '少年歌行风花雪月篇第' in filename and '集' in filename:
    if '第' in filename and '集' in filename:
        # 构建完整文件路径
        file_path = os.path.join(file_directory, filename)
        
        # 从文件名中提取集数
        episode_number = int(filename.split('第')[1].split('集')[0])
    
        # 读取Excel文件
        try:
            data = pd.read_excel(file_path)
            # 添加Episode列
            data['Episode'] = episode_number
            # 合并数据
            all_data = pd.concat([all_data, data])
        except Exception as e:
            print(f"Error reading {filename}: {e}")

# 重置索引
all_data.reset_index(drop=True, inplace=True)

# 按照 'Episode' 列进行升序排序
all_data.sort_values(by='Episode', inplace=True)

# 保存合并后的数据到一个新的Excel文件
output_file_path = os.path.join(file_directory, 'combined_data.xlsx')
all_data.to_excel(output_file_path, index=False)

print("All files have been combined and saved to 'combined_data.xlsx'.")
# 显示合并后的DataFrame的前几行
all_data.head(10)
```

##抖音微短剧数据

##注意：运行完成后需要将列名“评论内容”修改为“内容”

```python
import pandas as pd
import os

# 设置包含所有CSV文件的目录路径
file_directory = 'D:/研究生论文/小论文2/代码/01数据预处理/00data-初始-未合并集数/桃花马上请长缨'

# 列出所有CSV文件
csv_files = [file for file in os.listdir(file_directory) if file.endswith('.csv')]

# 初始化一个空的DataFrame来存储所有数据
all_data = pd.DataFrame()

for filename in csv_files:
    if '第' in filename and '集' in filename:
        # 构建完整文件路径
        file_path = os.path.join(file_directory, filename)
        
        # 从文件名中提取集数
        try:
            episode_number = int(filename.split('第')[1].split('集')[0])
        except ValueError as e:
            print(f"Error extracting episode number from {filename}: {e}")
            continue
        
        # 读取CSV文件
        try:
            data = pd.read_csv(file_path, encoding='utf-8-sig')
            # 添加Episode列
            data['Episode'] = episode_number
            # 合并数据
            all_data = pd.concat([all_data, data], ignore_index=True)
        except Exception as e:
            print(f"Error reading {filename}: {e}")

# 按照 'Episode' 列进行升序排序
all_data.sort_values(by='Episode', inplace=True)

# 保存合并后的数据到一个新的Excel文件（不加 encoding 参数）
output_file_path = os.path.join(file_directory, 'combined_data.xlsx')
all_data.to_excel(output_file_path, index=False, engine='openpyxl')

print("All files have been combined and saved to 'combined_data.xlsx'.")
# 显示合并后的DataFrame的前几行
print(all_data.head(10))
```
