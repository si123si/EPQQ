```python
import pandas as pd
import os

# è®¾ç½®åŒ…å«æ‰€æœ‰Excelæ–‡ä»¶çš„ç›®å½•è·¯å¾„
file_directory = 'D:/ç ”ç©¶ç”Ÿè®ºæ–‡/å°è®ºæ–‡2/ä»£ç /01æ•°æ®é¢„å¤„ç†/00data-åˆå§‹-æœªåˆå¹¶é›†æ•°/18æ‹©å¤©è®°ç¬¬äºŒå­£'   

# åˆ—å‡ºæ‰€æœ‰Excelæ–‡ä»¶
excel_files = [file for file in os.listdir(file_directory) if file.endswith('.xlsx') or file.endswith('.xls')]

# åˆå§‹åŒ–ä¸€ä¸ªç©ºçš„DataFrameæ¥å­˜å‚¨æ‰€æœ‰æ•°æ®
all_data = pd.DataFrame()

for filename in excel_files:
    #if 'å°‘å¹´æ­Œè¡Œé£èŠ±é›ªæœˆç¯‡ç¬¬' in filename and 'é›†' in filename:
    if 'ç¬¬' in filename and 'é›†' in filename:
        # æ„å»ºå®Œæ•´æ–‡ä»¶è·¯å¾„
        file_path = os.path.join(file_directory, filename)
        
        # ä»æ–‡ä»¶åä¸­æå–é›†æ•°
        episode_number = int(filename.split('ç¬¬')[1].split('é›†')[0])
    
        # è¯»å–Excelæ–‡ä»¶
        try:
            data = pd.read_excel(file_path)
            # æ·»åŠ Episodeåˆ—
            data['Episode'] = episode_number
            # åˆå¹¶æ•°æ®
            all_data = pd.concat([all_data, data])
        except Exception as e:
            print(f"Error reading {filename}: {e}")

# é‡ç½®ç´¢å¼•
all_data.reset_index(drop=True, inplace=True)

# æŒ‰ç…§ 'Episode' åˆ—è¿›è¡Œå‡åºæ’åº
all_data.sort_values(by='Episode', inplace=True)

# ä¿å­˜åˆå¹¶åçš„æ•°æ®åˆ°ä¸€ä¸ªæ–°çš„Excelæ–‡ä»¶
output_file_path = os.path.join(file_directory, 'combined_data.xlsx')
all_data.to_excel(output_file_path, index=False)

print("All files have been combined and saved to 'combined_data.xlsx'.")
# æ˜¾ç¤ºåˆå¹¶åçš„DataFrameçš„å‰å‡ è¡Œ
all_data.head(10)
```

##æ³¨æ„ï¼šè¿è¡Œå®Œæˆåéœ€è¦å°†åˆ—åâ€œè¯„è®ºå†…å®¹â€ä¿®æ”¹ä¸ºâ€œå†…å®¹â€

```python
import pandas as pd
import os

# è®¾ç½®åŒ…å«æ‰€æœ‰CSVæ–‡ä»¶çš„ç›®å½•è·¯å¾„
file_directory = 'D:/ç ”ç©¶ç”Ÿè®ºæ–‡/å°è®ºæ–‡2/ä»£ç /01æ•°æ®é¢„å¤„ç†/00data-åˆå§‹-æœªåˆå¹¶é›†æ•°/æ¡ƒèŠ±é©¬ä¸Šè¯·é•¿ç¼¨'

# åˆ—å‡ºæ‰€æœ‰CSVæ–‡ä»¶
csv_files = [file for file in os.listdir(file_directory) if file.endswith('.csv')]

# åˆå§‹åŒ–ä¸€ä¸ªç©ºçš„DataFrameæ¥å­˜å‚¨æ‰€æœ‰æ•°æ®
all_data = pd.DataFrame()

for filename in csv_files:
    if 'ç¬¬' in filename and 'é›†' in filename:
        # æ„å»ºå®Œæ•´æ–‡ä»¶è·¯å¾„
        file_path = os.path.join(file_directory, filename)
        
        # ä»æ–‡ä»¶åä¸­æå–é›†æ•°
        try:
            episode_number = int(filename.split('ç¬¬')[1].split('é›†')[0])
        except ValueError as e:
            print(f"Error extracting episode number from {filename}: {e}")
            continue
        
        # è¯»å–CSVæ–‡ä»¶
        try:
            data = pd.read_csv(file_path, encoding='utf-8-sig')
            # æ·»åŠ Episodeåˆ—
            data['Episode'] = episode_number
            # åˆå¹¶æ•°æ®
            all_data = pd.concat([all_data, data], ignore_index=True)
        except Exception as e:
            print(f"Error reading {filename}: {e}")

# æŒ‰ç…§ 'Episode' åˆ—è¿›è¡Œå‡åºæ’åº
all_data.sort_values(by='Episode', inplace=True)

# ä¿å­˜åˆå¹¶åçš„æ•°æ®åˆ°ä¸€ä¸ªæ–°çš„Excelæ–‡ä»¶ï¼ˆä¸åŠ  encoding å‚æ•°ï¼‰
output_file_path = os.path.join(file_directory, 'combined_data.xlsx')
all_data.to_excel(output_file_path, index=False, engine='openpyxl')

print("All files have been combined and saved to 'combined_data.xlsx'.")
# æ˜¾ç¤ºåˆå¹¶åçš„DataFrameçš„å‰å‡ è¡Œ
print(all_data.head(10))
```
